```python
from datetime import datetime

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# import seaborn as sns
```

## Análise Preliminar da Campanha

O objetivo inicial da campanhar é arrecadar R$ 700 milhões entre 27 de novembro de 2024 e maio de 2025. Alguns portais de notícias informam que a data final seria 10 de maio de 2025. Com base nessas informações, podemos calcular o valor diário (`daily_goal`) que deve ser arrecadado para que o objetivo seja atingido.

Se, ao final da campanha, o valor médio diário arrecadado for igual ou superior ao valor calculado, a campanha conseguiu atingir seu objetivo no prazo determinado. Por outro lado, se o valor for menor, então a campanha deverá ser estendida.

É importante observar que, da maneira como está sendo realizado, o cálculo da duração da campanha (`campaign_duration`) não leva em consideração o primeiro dia da campanha.

```python
# Objetivo de arrecadação da campanha
fundraising_goal = 7 * 10**8  # R$ 700 milhões

date_fmt = "%d/%m/%Y"
date_begin = datetime.strptime("27/11/2024", date_fmt)
date_end = datetime.strptime("10/05/2025", date_fmt)

# Número de dias entre o início e o fim da campanha (duração)
campaign_duration = (date_end - date_begin).days  # + 1 (opcionalmente)

daily_goal = fundraising_goal / campaign_duration
print(f"Objetivo diário de arrecadação: R$ {daily_goal:.2f}.")
```

## Análise dos Dados da Campanha

Vamos carregar os dados da campanha. O arquivo em formato CSV contém uma tabela com três colunas:

1. `valor`: valor total arrecadado;
2. `data`: data (formato "dd/mm/aaaa"); 
3. `horario`: horário (formato "hh:mm").

Algumas observações sobre o conjunto de dados são importantes: 

- devido a instabilidades no site, principalmente no início da campanha, alguns dados foram inseridos manualmente com base em capturas de tela encontradas na Internet;
- os dados não são disponibilizados em intervalos regulares de tempo, ou seja, o intervalo de tempo entre um dado e outro varia.

```python
csv_url = "https://raw.githubusercontent.com/ptrckfc/dados-arrecadacao-doe-arena-corinthians/refs/heads/main/arrecadacao_DoeArenaCorinthians.csv"
campaign_df = pd.read_csv(csv_url, header=0)

campaign_df.info()
```

### Evolução da Campanha

A tabela construída contém as seguintes informações:

1. `data`: data (formato "dd/mm/aa");
2. `valor_arrecadado`: valor total arrecadado (cumulativo);
3. `valor_objetivo`: valor que deveria ter sido arrecadado, na média, para que a campanha consiga atingir seu objetivo no prazo estabelecido;
4. `valor_diario`: valor arrecadado por dia;
5. `arrecadado_objetivo`: relação entre o valor total arrecadado e o valor objetivo.

Vamos considerar $T(t)$ o valor total arrecadado até o dia $t$. O valor diário, $v(t)$, é obtido fazendo:

$$
v(t) = T(t) - T(t -1)\text{.}
$$

Vamos representar o valor objetivo por $O(t)$. A relação, $r(t)$, entre o valor total arrecadado e o valor objetivo é dada por:

$$
r(t) = \frac{T(t)}{O(t)}\text{.}
$$

```python
campaign_days = campaign_df["data"].unique()

data = list()

# Valores são divididos por 10^6 (apenas para trabalharmos com valores menores)
for campaign_day in campaign_days:
    date = datetime.strptime(campaign_day, date_fmt)
    fundraised = campaign_df[campaign_df["data"] == campaign_day].iloc[-1]["valor"] / 10**6
    goal = daily_goal * (1 + (date - date_begin).days) / 10**6    
    data.append([campaign_day, fundraised, goal])

daily_df = pd.DataFrame(data, columns=["data", "valor_arrecadado", "valor_objetivo"])

# Cálculo do valor diário arrecadado e da relação arrecadado-objetivo
daily_df["valor_diario"] = daily_df["valor_arrecadado"].diff(periods=1)
daily_df["arrecadado_objetivo"] = daily_df["valor_arrecadado"] / daily_df["valor_objetivo"]

# Remove o dia 26/11/2024 da base de dados (anterior ao início da campanha)
daily_df.dropna(axis=0, inplace=True)

daily_df.head(10)
```

```python
import math

fig, ax = plt.subplots()

ax.plot(
    daily_df["data"],
    daily_df["valor_objetivo"],
    marker="o",
    linestyle="dashed",
    color="black",
    alpha=0.5,
    label="Objetivo"
)

sct_ax = ax.scatter(
    daily_df["data"],
    daily_df["valor_arrecadado"],
    c=daily_df["valor_diario"],
    sizes=daily_df["valor_diario"] * 100,
    cmap="jet_r",
    alpha=0.8,
    label="Arrecadado"
)

fig.colorbar(sct_ax, aspect=50)

ax.set_title("Evolução da Campanha")
ax.set_xlabel("Data")
ax.set_ylabel("Arrecadação (em milhões de R$)")
ax.legend()

ax.minorticks_on()
ax.grid(visible=True, which="both", alpha=0.3, linestyle="--")

fig.autofmt_xdate()
```

```python
fig, ax = plt.subplots()

ax.stem(
    daily_df["data"],
    daily_df["arrecadado_objetivo"]
)

ax.set_xlabel("Data")
ax.set_ylabel("Relação")
ax.set_title("Relação entre o valor total arrecadado e o valor objetivo")

ax.minorticks_on()
ax.grid(visible=True, which="both", alpha=0.3, linestyle="--")

fig.autofmt_xdate()
```

#### Percentagens

```python
daily_df.head()

percentage_df = pd.DataFrame(columns=["percentagem_total", "percentagem_objetivo", "percentagem_diaria"])

percentage_df["percentagem_total"] = daily_df["valor_arrecadado"] / 7
percentage_df["percentagem_objetivo"] = daily_df["valor_objetivo"] / 7
percentage_df["percentagem_diaria"] = daily_df["valor_diario"] / (daily_goal / 10**8)

percentage_df.head(10)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 2.7), layout="tight")

axs[0].plot(daily_df["data"], percentage_df["percentagem_total"], marker="o", label="Arrecadado")
axs[0].plot(daily_df["data"], percentage_df["percentagem_objetivo"], linestyle="--", label="Meta")

axs[0].set_xlabel("Data")
axs[0].set_ylabel("Percentagem (%)")
axs[0].set_title("Percentagem total")
axs[0].legend()

axs[0].minorticks_on()
axs[0].grid(visible=True, which="both", alpha=0.3, linestyle="--")

axs[1].plot(daily_df["data"], percentage_df["percentagem_diaria"], marker="o", label="Arrecadado")
axs[1].plot([daily_df["data"].iloc[0], daily_df["data"].iloc[-1]], [100, 100], linestyle="--", label="Meta")

axs[1].set_xlabel("Data")
axs[1].set_title("Percentagem diária")
axs[1].legend()

axs[1].minorticks_on()
axs[1].grid(visible=True, which="both", alpha=0.3, linestyle="--")

fig.autofmt_xdate()
```
